```python
def f1(dataset, x, y, mr):
    row1 = dataset.loc[dataset['spatial_mr_rate_perc'] == y]
    row2 = row1.loc[dataset['tensor_dim'] == x]
    row3 = row2.loc[dataset['mr'] == mr]
    result = row3['tcs_cost']
    return result
```
